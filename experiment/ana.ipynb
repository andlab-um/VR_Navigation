### 程序随机化

1. 随机10个x1, x2 以构建20条数据
2. 表格的分类： 
3. 一共 6个 blocks 这里我们固定每个人相同试次顺序


```python
# col1: location of the figures
# col2: Exists? 'S' for True and 'K' for False 
# col3: Things or Buildings?


# 修改代码： 

import pandas as pd

def gen_table(x1,x2):
    np.random.shuffle(x1)
    np.random.shuffle(x2)
    x1 = x1[:20]
    x2 = x2[:20]
    data = {'fig_location': x1 + x2, 'corr': ['k']*len(x1) + ['s']*len(x2)}

    indicator = []
    for item in data['fig_location']:
        if 'Avatar' in item:
            if 'Case' in item:
                indicator.append('Case')
            else:
                if 'Female' in item:
                    indicator.append('Female')
                else:
                    indicator.append('Male')
        else:
            number = item.split('/')[-1].split('_')[0]
            if int(number) <= 10:
                indicator.append('scene_things')
            else:
                indicator.append('building')

    data['indicator'] = indicator
    
    df = pd.DataFrame(data)
    df_randomized = df.sample(frac=1).reset_index(drop=True)
    
    return df_randomized


def save_excel(data: pd.DataFrame,sheet_name: str):

    data.to_excel('orderexcel/'+sheet_name, sheet_name='Sheet1', index=False)

    


for i in range(6):
    if i % 2 == 0:
        data = gen_table(X_dis_case_scene,X_case_scene)
    else:
        data = gen_table(X_dis_avatar_scene,X_avatar_scene)
    save_excel(data, str(i+1)+'.xlsx')
```
